### Productos (toda la info junta)

```python
soup = BeautifulSoup(response.content)
```

```python
productos = soup.find_all('div', class_= 'product-container')
```

```python
productos
```

```python
df_productos = pd.DataFrame.from_records(productos)
df_productos
```

```python
df_productos.to_csv("productos.csv")
```

### Nombre de los productos (títulos)

```python
productos2 = soup.find_all('h5', class_= 'product-name')
```

```python
productos2
```

```python
df_productos2 = pd.DataFrame.from_records(productos2)
df_productos2
```

```python
df_productos2.to_csv("productos2.csv")
```
